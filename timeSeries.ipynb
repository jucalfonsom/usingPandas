```python
import pandas as pd
import numpy as np
```

# Load covid 19 dataset from Kaggle

https://www.kaggle.com/sudalairajkumar/novel-corona-virus-2019-dataset?select=covid_19_data.csv


```python
df = pd.read_csv('db/covid_19_data.csv')
df
```

## Convert ObservationDate column to datetime

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df
```

```python
list(df)
```

## Filter dataset

```python
df_time = df[[
 'ObservationDate',
 'Country/Region',
 'Confirmed',
 'Deaths',
 'Recovered']].groupby('ObservationDate').sum()

df_time
```

## Create different dataframes

```python
df1 = df_time['Confirmed'].iloc[10:15]
df1
```

```python
df2 = df_time['Confirmed'].iloc[12:17]
df2
```

```python
df1-df2
```

# Calculate variation by day

```python
df_time.diff().head(10)
```

## Replace NaN values on 2020-01-22

```python
df_time.head(1).to_dict()
```

```python
df_diff = df_time.diff().fillna({'Confirmed': 555.0,
 'Deaths':  17.0,
 'Recovered': 28.0})

df_diff
```

## Acumulative sum

We can get the original dataframe with Acumulative sum

```python
df_diff.cumsum()
```

## Statistical estimates

```python
# 7 days

df_diff.resample('7D').sum()
```

```python
# Every week on sunday

df_diff.resample('W-Sun').sum()
```

```python
# Every month

df_diff.resample('M').sum()
```

```python
df_diff.resample('M').count()
```

```python
df_diff.resample('M').mean()
```

# NaN or empty values in time series

```python
#12 hours

df_diff.resample('12h').sum()
```

```python
df_diff.resample('12h').mean()
```

```python
df_cum = df_time.resample('12h').sum(min_count = 1)
df_cum
```

```python
df_cum.bfill()
```

```python
df_cum.ffill()
```

```python
df_cum.fillna(-1000)
```

```python
df_cum = df_cum.interpolate()
df_cum
```

## Create new column

```python
df_cum['rate'] = 1- df_cum['Deaths'] / df_cum['Confirmed']
df_cum
```

## How to extract information when timestamp variable is not the index

```python
df_cum = df_cum.reset_index()
df_cum
```

```python
#Groupby for time series

df_cum.groupby(pd.Grouper(key = 'ObservationDate', freq = 'M'))[['rate']].mean()
```

## Apply function rolling (time window)

```python
import matplotlib.pyplot as plt
```

```python
sr = df_cum.groupby(pd.Grouper(key = 'ObservationDate', freq = '1D'))['rate'].mean()
sr
```

```python
plt.plot(sr)
```

```python
plt.plot(sr)
plt.plot(sr.rolling(window = 7).mean())
plt.plot(sr.rolling(window = 14).mean())
```

```python
plt.plot(sr.rolling(window = 14).apply(lambda x : np.std(x)))
```
