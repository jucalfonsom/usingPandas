```python
import seaborn as sns
```

# Load tips dataset

```python
df = sns.load_dataset('tips')
df
```

# Adding new column

```python
df['ones'] = 1
df
```

# Group by

```python
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum()
df_g
```

```python
# level = 0: sex
# Convert values to percentage

df_g.groupby(level = 0).apply(
    lambda x : x / x.sum() * 100
)
```

# Convert numerical to categorical variable

Creating intervals

```python
pd.cut(df['total_bill'], bins = 3).value_counts()
```

```python
# Specific intervals

pd.cut(df['total_bill'], bins = [3, 18, 35]).value_counts()
```

```python
# Specific intervals

pd.cut(df['total_bill'], bins = [3, 18, 35]).value_counts()
```

```python
pd.cut(df['total_bill'], bins = [3, 18, 35, 60]).value_counts()
```

```python
df['bin_total'] = pd.cut(df['total_bill'], bins = [3, 18, 35, 60])
df
```

## Group by new categories

```python
df.groupby(['time', 'bin_total'])[['ones']].count().groupby(level = 0).apply(
    lambda x : x / x.sum() * 100
)
```
